```python
# converting the data type of the column 'education' into numeric values 
data["education"] = pd.to_numeric(data["education"], errors='coerce')
data = data.dropna()
```

```python
data['education'].dtypes
```

```python
x = data.drop("TenYearCHD", axis=1)
x.columns
```

```python
y = data['TenYearCHD']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split( x, y,test_size= .3
                                                 ,random_state= 42 ,
                                                 stratify= y )
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
print(data.shape)
```

```python
x_train['education']
```

```python
# scaling the data using MinMaxScaler 
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
# use the logistic regression classifier to train the data and test the model 
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
model = lg.fit(x_train_scaled,y_train)
y_predict = model.predict(x_test_scaled)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Make predictions on the test data
y_pred = model.predict_proba(x_test_scaled)[:,1]

# Calculate the fpr, tpr, and thresholds for the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred)

# Calculate the AUC
roc_auc = roc_auc_score(y_test, y_pred)

# Plot the ROC curve
plt.plot(fpr, tpr, label='AUC = {:.3f}'.format(roc_auc))
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
```

```python
# using the classificaion report to evaluate the model 
from sklearn.metrics import classification_report

report = classification_report(y_test,y_predict)
print(report)
```
